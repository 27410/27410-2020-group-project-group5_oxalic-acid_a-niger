# Oxalic acid production with Aspergillus niger

## 1. Introduction

Oxalic acid has been of industrial importance ever since, traditionally obtained through chemical synthesis. However, the chemical synthesis affords the use of harsh chemicals like sulphuric acid and high temperatures up to 360 °C leading to a high energy consumption. Biological production of oxalic acid could provide a sustainable alternative with the possibility of using waste products like cashew apple juice as substrate [1]. However, major drawbacks are low yields and low productivity, mainly due the oxidation of glucose and loss of energy in unfavorable pathways [1].
The overall aim of this project is to evaluate engineering strategies for the production of oxalic acid in the cell factory Aspergillus niger.

### 1.1 Literature review of the compound
Oxalic acid is a dicarboxylic acid with a molecular weight of 90,03 g/mol. The molecule resembles ethane with substitutions of carboxyl groups at positions 1 and 2 [2].
Oxalic acid has several applications. The most common one is contribution in laundry detergents as the acid rinse, being responsible for converting insoluble iron compounds into soluble complex ions and thus removing dust such as rust and ink stains or removing scale from car radiators [3].
It is also applied for bleaching clothes and in medicine it is used for further purifying chemicals or to dilute them [4].



Figure 1: Structure of oxalic acid [5]

With the help of the Kegg database, oxalic acid could be found in the glyoxylate and dicarboxylate metabolism of Aspergillus niger. There, two main routes to produce the metabolite oxalic acid could be identified. Both routes are branch-offs from the TCA cycle. The first one is starting from the TCA-cycle intermediate oxaloacetate, which is reduced to another intermediate of the TCA-cycle: Malate. This reaction is an anti-clockwise reaction of the TCA cycle. Thereafter, Malate is converted into Glyoxylate by reacting with CoA. Then, this Glyoxylate is oxidized into the desired oxalate, which is catalyzed by an oxygen oxidoreductase. 

The other route starts two stages later in the TCA cycle going from oxaloacetate to Citrate and then to Isocitrate, which is then split into Succinate and Glyoxylate which is mediated by the enzyme isocitrate glyoxylate-lyase. 
As mentioned above in the other route, the Glyoxylate is finally oxidized into oxalate [6].


The global market size for oxalic acid has been around 620 million USD in 2017 [7]. In 2019 the market was assumed to grow by roughly 177 million USD in the following year [8]. Driver for  a higher demand in the market is mostly China, but followed by Japan, US and Germany [8]. Depending on the desired purity, prices for oxalic acid range from 0,5 - 1 USD/kg [9]. The price always hints towards the challenge of introducing a biotechnologically competing process: The product price is not much higher than the substrate cost (~0,1-0,2 USD/kg), therefore a cheap and efficient process is required [10].


### 1.2 Literature review of the cell factory
A. Niger counts itself among the most important biotechnological microorganisms. A. Niger has already been used biotechnologically for different products since many decades, so that much information about the metabolism and fermentation demands are already available . Moreover, it is well-known for the extracellular production of enzymes, mostly food enzymes, and citric acid in high yields.[6] This ability to secrete its products in high concentration into the fermentation medium can be considered as an advantage as well, since downstream costs are reduced as a cell disruption method is not required [11].
A. Niger is used for biotransformations and has been utilized in that regard for the over-expression of food enzymes [11].
Another advantage is that many products from A. Niger are characterized as GRAS products by the US-FDA [11].

So, a number of several protease genes containing a secretion peptide could be identified, which pose a risk for product degradation in industrial fermentation. Moreover, genes were discovered for the production of the carcinogenic mycotoxins fumonisin and ochratoxin A. Even though a draft genome of A. Niger was published, only half of the discovered 14.165 ORF’s were functionally predictable. This ‘Black Box’ poses an issue to get a better understanding on the level of systems biology.

There have been some studies about oxalic acid production in Aspergillus niger. An older study reported a product concentration of 33,8 g/l and a productivity of 4,1 g/l*d. Finally, they obtained a yield of 0,3 kg oxalic acid/kg sucrose [12]. More recent studies identified an oxalic acid overproducing mutant with a productivity of approximately 3 g/l*d and a final titer of 11 g/l. Highlight of this study is the oxalic acid overproducing phenotype, which had an increased yield of glucose on oxalic acid of 90 % compared to the wildtype. This was achieved by knocking out pH responding transcription factors. Therefore, manipulating transcription factors could be a solution to debottleneck production of oxalic acid in Aspergillus niger [13].
In a more recent study researchers overexpressed the gene oahA, which is encoding the enzyme oxaloacetate hydrolase (EC 3.7.1.1). The overexpression leads to an increase of concentration from 15,6 g/l to 28,9 g/l compared to the unmodified strain. Finally, they reported a yield of 64,2 % of the maximum theoretical yield for glucose to oxalic acid . Therefore, an overexpression of that gene could further increase flux towards oxalic acid [14]. In a even more recent paper, a titer of 123 g/l oxalic acid was reached and a productivity of 17 g/l*d [1].
In general, Aspergillus niger seems to be a promising cell factory for this compound.

## 2. Problem definition

As an organic acid, oxalic acid can be considered as a low value chemical, meaning having an abundant and cheap feedstock as well as great fermentation metrics is key to compete with the existing chemical processes. Therefore, as a rule of thumb, a final product titer of minimum 100 g/L at the end of fermentation is desired, which has already been achieved in Aspergillus niger fermentations by some research groups. However, the productivity has to be increased to at least 5 g/L*h. Productivity has, as presented previously already, been significantly lower than the desired value. Finally, the yield YSP has to be improved closer to the maximum theoretical yield considering sucrose as substrate. Highest achieved yield compared to the maximum theoretical yield was 64,2 %.

To achieve an economically favourable process, different substrates and pH and gene knockouts, should be investigated with an optimized Aspergillus niger iJB1325 model [15].

The yield of substrate on the product seems to be a major problem. Sucrose and also lactose were separately chosen as a carbon source in the reference paper [16]. Thereofore, their suitability for favoring oxalate production needs to be assessed. 
However, especially for sucrose, this had the drawback of getting degraded into glucose at the beginning of the fermentation, which had an unfavorable consumption of the glucose and flux into gluconic acid as a consequence. Therefore, identifying the sucrose-degrading extracellular enzyme and knocking it out could be a promising approach. Since this could probably have a negative effect on the growth rate of the host as this enzyme could be essential for a specific pathway, an alternative approach could be to prevent the unfavorable flux of towards gluconic acid by knocking out genes involved in expressing the enzymes involved in the reactions of the pathway. 

## 3. *If Project category II:* Selection and assessment of existing GSM
The model chosen for optimizing the production of oxalic acid in Aspergillus niger iJB1325 was chosen based on its date of publication and scope. This model resembles the newest genome-based model for Aspergillus niger, it was published in 2018. The given model is an updated version of the model iMA871 which was published in 2009. In general, the overall coverage of the genome-scale model and the assignment of genes to reactions was improved.  Special emphasis was also put on including modeling of recently discovered secondary metabolites and other proteins [15]. In total, the chosen model  iJB1325 extends the model iMA871 by 940 reactions, 777 metabolites and 454 genes resulting in a total number of 2320 reactions, 1818 metabolites and 1325 genes. Overall, the chosen model is considered the best choice because of having integrated big amounts of recent data which enables trustful simulations of flux distributions [15].
Moreover, the GSM has been updated based on including data from literature, patents and several databases like BRENDA for instance.
On top of that, two further models of A.niger have been partly integrated in the updated model iMA871, one is a de novo-generated based model on an advanced automated method , and one was published by Lu et al [17].

As part of the model  iJB1325, 471 test cases have been implemented which can be run when updating the model, in order to evaluate if the different sources of data on which the model was built go hand in hand. Those tests entail presence/absence of growth on different C-and N-sources (371 cases), presence/absence of knockout-mutants (73 cases), overall system check like producing biomass precursors.
Performing all tests yields a score of 79 % passing and 21% failing the tests. Most of those failures were due to unknown metabolic pathways (76,5 %), inconsistencies between different experiment data (15,3 %) and undefinable  reasons (8,2 %).
The model is furthermore backed up with evidence from literature in different levels. All in all, 19 % of the reactions which are included in the model have direct experimental back-up, whereas 17 % of the reactions have a gene assigned which has strong similarities to genes encoding for certain characterized enzymes in closely related species, mostly A. nidulans and A.fumigatus. The remaining reactions were not able to be assigned to a gene or are taken over from the model iMA871 as best scores resulting from sequence similarity studies of characterized enzymes or domains.
The date of publication, extended number of reactions,metabolites and genes (in relation of  iMA871) as well as the presence of validation tests and evidence from literature make this model a well-suited approach to perform metabolic studies for the production of oxalic acid in A. niger. The memote analysis of the GSM-model revealed that the model is excellent in consistency, but lacking metabolites, reactions, gene and SBO annotations. A total score of 46 % was reached [18]. The full report is supplemented in the repository as Memote full report.

# 4.6 Dynamic Flux Balance Analysis (dFBA)

- ## Simulation of Batch cultivations under different medium and aeration conditions

In order to simulate A.niger growth in a batch process when producing oxalic acid over a period of time, we had to employ a dFBA approach. To properly comprehend A.niger growth, different conditions in terms of substrates and aeration have been applied which led to different results which will be explained in the following:

Furthermore, it is important to note that when simulation a dFBA certain criterias have to be taken into account:

   -**DFBA model**: our model is build with the following ODE system 
  
 \begin{align}
    \frac{dX}{dt} &= v_{BM}X \\
    \frac{dC_j}{dt} &= v_jX
\end{align}

where $X$ is the Biomass (_gDW_); $v$ is the flux of a exchange reaction $j$ (_mmol/gDW/h_) or the exchange biomass reaction $BM$ (_g/gDW/h_, to measure growth). 

   -**Kinetic variables**: 
  
For this example, the metabolites $j$ in the medium will be **Glucose/Sucrose/Lactose, Oxygen, Oxalic acid** and a **Biomass** variable.

   -**Exchange fluxes**:
   
We make exchange fluxes to appear in the simulation, which is corresponding to the exchange reactions of the cobrapy model. 

   -**Rhs expressions**: We calculate symbolic expressions for the time derivative of each kinetic variable currently in the model. 

   -**Lower/upper bound expressions**:We add symbolic expressions for calculating lower/upper bounds of selected exchange fluxes
currently in the model. It is important to note that the lower bounds will follow a Michaelis-Menten kinetics:

$$
lb = - \frac{C_jV_{max}}{ C_j + K_m}
$$

Kinetic parameters:  $V_{max}$= 15 [mmol gDW^-1 h^-1] and $K_m$= 0.0024 [mmol/L O2] --> Michaelis-Mentent constant

   -**Initial conditions and simulation**: Initial conditions are added to represent the desired simulation. Concentrations and trajectories of kinetic variables will be returned.
   
   dfba_model.add_initial_conditions(
    {
        "Biomass": 0.05, # (gDW/L)
        "Sucrose": 13, # (g/L)
        "Oxygen": 0.2,  # (mmol/L)
        "Oxalic acid": 0.0,  # (g/L)
    }

   -**Plotting results**: Two graphs will appear; one simulating metabolites concentrations and the other metabolites trajectories.
   


Firstly, to be able to better analyse the model we set a minimum bound for the biomass reaction in our dfba model. In this case, we set it to 30% of the amount of flux that goes through the biomass reaction. In addition, the objective of our dfba model is set to be the exchange reaction for oxalic acid, the product of interest.

# 5. Discussion

## 5.2 Discussion of Maximum Theoretical Yields and Medium Characterization 
In order to understand our model, a computational analysis was developed to calculate yields, production,growth and characterize substrate medium for our host system, A.niger, during production of Oxalic acid with a sucrose based carbon source. The result of oxalic acid production in A.niger with sucrose is 59.99 [mmol gDW^-1 h^-1] and the maximum theoretical yield of oxalic acid on sucrose in A.niger is 5.99 [mmol-Oxal / mmol-suc]
Results show that A.niger growth, oxalic acid production and max.theoretical.yield values all double when employing sucrose/lactose as a carbon source under aerobic conditions. This implies that the carbon substrate could be further investigated to employ a cheaper carbon source, that could produce higher growth rates and increase oxalic production.This could be done by evaluating their max.theoretical.yields, gowth and production rates.
In addition, results show that when disrupting S,P,N metabolites under aerobic conditions A.niger growth is stopped. This signifies the essentiality of these components and opens the possibility to again employ cheaper components to optimize the system.
Finally, as expected from literature work, results show no growth when working under anaerobic conditions. This results strenghten the known fact that Aspergillus species are highly aerobic. Nevertheless, further research could be done to produce more anaerobic-tolerant strains.
## 5.3 Discussion of Flux Analysis 
In general, two major pathways were identified for being able to produce oxalate. Those were represented by the reaction r51 and the manually added reaction GOOR, respectively.  
By comparing those two possible routes to oxalate, it is apparent that r51 is the more favorable reaction in terms of flux, regardless of which carbon source was used in the media (glucose, lactose, sucrose). Even though r51 was the favorable reaction in flux, the manually reaction GOOR showed a higher sensitivity upon switching the carbon source from glucose to sucrose or lactose and doubled its flux, whereas r51 increased by 21%. Thus, the flux through the pathway of citric acid towards oxalate (=GOOR) seems to be more affected by the change of carbon source than r51.
In the flux analysis the findings of the reference paper that sucrose and lactose are shown to be favorable for oxalic acid production are confirmed. Still, it is surprising that the fluxes for the target reactions r51 and GOOR obtained with sucrose or lactose as the carbon source do not differ from one another. It is really unlikely and requires further investigation of the model. 

## 5.4 Discussion of knockout strategy
First, the genes coding for the enzyme oxaloacetate acetylhydrolase, which catalyzes the target reaction r51 were investigated. Here, only when the two identified genes were knocked-out simultaneously the flux was 0, whereas a knockout of just one of the genes did not change the flux at all. This shows that the two genes code for the entire enzyme and yield sufficient amounts of enzyme, which hints towards the fact that the fungus has the two genes as an evolutionary back-up strategy, if one of the genes loses its function due to mutations.

Then, the knockout of the invertase was investigated, which was identified as the enzyme which is responsible for the unfavorable cleavage of sucrose to glucose in the beginning of the fermentation, a problem discussed by the reference paper. Knocking out the gene had neither any influence on the growth rate of the fungus nor on the flux of the main target reaction r51, which is against all expectations. So, one explanation that nothing changes is that there are alternative sucrose-degrading enzymes, which were not able to be identified in the model. This is supported by the fact that the growth is not influenced by the knockout of the invertase. In conclusion, this knockout is not recommendable.

Another approach was to knock out the genes which are coding for the enzyme which is responsible for the fast oxidation of glucose to gluconic acid. This is another issue discussed in the reference paper, since glucose is “wasted” to produce oxalic acid. Therefore, the two genes for the identified reaction r164a were knocked out, at first respectively, and thereafter combined.  Also here, knocking out the genes neither respectively nor combined did affect the flux of the target reaction r51 nor the growth of the fungus. Thus, also this knockout is not having any effect, which is again against the expectations since the flux should be redirected in other pathways, like r51 for instance. 
Also that the growth rate is not influenced by the knockout is surprising, since this is a reaction included in the pentose-phosphate shunt which is known to be necessary for producing NADPH or building blocks for nucleotides (RNA and DNA) which are essential for the survival of the fungus. 

Finally, OptGene and OptKnock were performed to suggest possible combinations of mutations to favor the oxalate-formation. Both analysis methods did not yield any solution. So, apparently the analysis tools are not applicable here which can have several reasons. Further investigation is required.

In addition, lucose concentrations do not modify oxalic acid production.

## 5.6 Dynamic Flux Balance Analysis (dFBA)
In order to investigate our host system growth when producing oxalic acid in a batch process, we had to employ a dFBA approach. For this approach different initial conditions of substrates and aeration were tested. The resulting simulations show how employing sucrose/lactose instead of glucose increases the oxalic acid production curve. In addition, when studying metabolites trajectories, the oxalic acid flux increases when it is being produced while oxygen and sucrose fluxes decrease. In essence, these batch simulations showcase how production of oxalic acid by A.niger can be optimized by employing sucrose/lactose as the starting carbon source substrate in detriment of glucose. This dFBA approach could be further developed to analyse system variations when utilizing new metaboilites and conditions.








## 6. Conclusion

Changing the substrate from glucose to lactose or sucrose resulted in an increased growth rate and product flux which is confirming the findings of experimental studies conducted. Also, a knockout of the invertase did not improve the flux of the target reactions. Deletion of genes encoding enzymes for the competing pathway towards gluconate resulted in no change in growth rate and product flux either. The use of OptGene and OptKnock did not yield any interpretable data.

The aim of this project was to increase productivity and yield of oxalate. Those metrics have been achieved only with performing medium changes, whereas all knockout strategies failed to achieve that goal. Further research could be conducted in terms of performing over-expression of the sucrose transporter for instance to prevent the early sucrose degradation mediated by sucrose-degrading enzymes. Moreover, the pH of the medium could be shifted to a value > 6 since it was reported to have a slight beneficial effect on oxalate productivity.

In conclusion, all findings from the reference paper in terms of media composition could be confirmed, whereas strategies to overcome the mentioned issues of sucrose degradation in the beginning of fermentation and unfavorable oxidation towards gluconate, were not successful.
